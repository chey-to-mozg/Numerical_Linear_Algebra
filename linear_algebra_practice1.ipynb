### First part

#### what are the constants $C_1$ and $C_2$ such that $C_1 \|x\|_{2} \leq \|x\|_{\infty} \leq C_2 \| x\|_{2}$ for any vector $x$

- $\|x\|_{2} = \sqrt{\sum_{i=1}^{n}{|x_i|^2}}$, where $n -$ length of the vector $x$

- $\|x\|_{\infty} = \max{|x_i|}$

- Let's look at $C_1 \|x\|_{2} \leq \|x\|_{\infty}$. $C_1 \leq \frac{\|x\|_{\infty}}{\|x\|_{2}}$.
The maximum "distance" between these norms will be when all $x_i$ are equivalent and non-zero.
Let value of all $x_i = x`$. Then $C_1 \leq \frac{x`}{\sqrt{n{|x`|^2}}} =  \frac{1}{\sqrt{n}}$.

$C_1 = \frac{1}{\sqrt{n}}$

- Let's look at $\|x\|_{\infty} \leq C_2 \| x\|_{2}$. $C_2 \geq \frac{\|x\|_{\infty}}{\| x\|_{2}}$.
As $\|x\|_{2} \geq \|x\|_{\infty}$ (for example, $\sqrt{|x_1|^2 + |x_2|^2} > max(|x_1|, |x_2|)$),
the highest value of $\frac{\|x\|_{\infty}}{\| x\|_{2}}$ obtain when vector $x$ contain only one non-zero element.
Then $C_2 \geq \frac{x`}{\sqrt{|x`|^2}} = 1$.

$C_2 = 1$

#### Prove that $\| U A \|_F = \| A U \|_F = \| A \|_F$ for any unitary matrix $U$.

$\| U A \|_F = \sqrt{trace((UA)^*UA)} = \sqrt{trace(A^*U^*UA)} = \sqrt{trace(A^*A)} = \| A \|_F$

$\| A U \|_F = \sqrt{trace((AU)^*AU)} = \sqrt{trace(U^*A^*AU)} = \sqrt{trace(UU^*A^*A)} = \sqrt{trace(A^*A)} = \| A \|_F$


#### Prove that $\| U A \|_2 = \| A U \|_2 = \| A \|_2$ for any unitary matrix $U$.

$\| U A \|_2 = sup_{x \ne 0} \frac{\| U A x \|_2}{\| x \|_2}$. $Ax$ - vector and from unitary
invariance of vectors $\| U z \|_2 = \| z \|_2$.

Then, $sup_{x \ne 0} \frac{\| U A x \|_2}{\| x \|_2} = sup_{x \ne 0} \frac{\| A x \|_2}{\| x \|_2} = \|A \|_2$

$\| A U \|_2 = sup_{x \ne 0} \frac{\| A U x \|_2}{\| x \|_2}$. Let $z = Ux$, then $x = U^*z$

Then, $sup_{x \ne 0} \frac{\| A U x \|_2}{\| x \|_2} = sup_{z \ne 0} \frac{\| A z \|_2}{\| U^* z \|_2} = sup_{z \ne 0} \frac{\| A z \|_2}{\| z \|_2} =  \|A \|_2$

$\| A \|_2 = \| U \Sigma V \|_2 = \| \Sigma \|_2 = \sigma_1(A)$

Let $\sigma_1(A) = ... = \sigma_r(A)$, then $\| \Sigma \|_F = \sqrt{\sum_{i=1}^r{\sigma_1(A)}} = \sqrt{r} \sigma_1(A)$

### First part
$$ F(U, V) = \frac{1}{2}\|X - UV\|_F^2 $$

#### Derive analytical expression for the gradient of the function $F$ with respect to $U$

Let's look at $d(\| X \|_F^2)$

$d(\| X \|_F^2) = d( (X,X) ) = \{ d(x) = 1 dx \} = d(X,X) = 2(X, dX)$

$dF(U) = d( \frac{1}{2}\|X - UV\|_F^2 ) = (X - UV, dU(-V)) = trace((X - UV)^T dU(-V)) = trace((-V)(X - UV)^T dU) = ((X-UV)(-V)^T, dU)$

$\nabla F(U) = (X-UV)(-V)^T $


#### Derive analytical expression for the gradient of the function $F$ with respect to $V$

$dF(V) = d( \frac{1}{2}\|X - UV\|_F^2 ) = (X - UV, -UdV) = trace((X - UV)^T(-U) dV) = ((-U)^T(X-UV), dV)$

$\nabla F(V) = (-U)^T(X-UV) $


#### Estimate computational complexity of computing these gradients (in big-O notation)

Substracting complexity - $O(n^2)$

Multiplication complexity in common case - $O(n^3)$

Transpose complexity - $O(1)$ (change order of indexes)

$O((-U)^T) = O((-V)^T) = O(1)$

$O(UV) = O(n^2 k)$

$O(X - UV) = O(n^2 k + n^2)$

$O((-U)^T(X-UV)) = O((X-UV)(-V)^T) = O(1 + n^2k + n^2 + n^2k) = O(2n^2k + n^2 + 1) = O(n^2k)$

##### Compare timing of analytical computations versus timing the automatic differentiation with JAX.

````python
import jax
import jax.numpy as jnp
from jax.config import config
import time
config.update('jax_enable_x64', True)

@jax.jit
def f(X, U, V):
  res = (jnp.linalg.norm(X - U @ V) ** 2) / 2
  return res

def myGradU(X, U, V):
  return (X - U @ V) @ (-V).T

def myGradV(X, U, V):
  return (-U).T @ (X - U @ V) 
````

````python
n = 5000
k = 4000

rng = jax.random.PRNGKey(322)

X = jax.random.normal(rng, (n, n))
U = jax.random.normal(rng, (n, k))
V = jax.random.normal(rng, (k, n))

autogradfU = jax.grad(f, argnums=1, has_aux=False)
jit_gradU = jax.jit(autogradfU)
autogradfV = jax.grad(f, argnums=2, has_aux=False)
jit_gradV = jax.jit(autogradfV)

# print(f'jax grad U = {autogradfU(X, U, V)}')
# print(f'my  grad U = {myGradfU}')
# print(f'jax grad V = {autogradfV(X, U, V)}')
# print(f'my  grad V = {myGradfV}')

print(f'check correctness U: {jnp.linalg.norm(jit_gradU(X, U, V) - myGradU(X, U, V))}')
print(f'check correctness V: {jnp.linalg.norm(jit_gradV(X, U, V) - myGradV(X, U, V))}')

print("My grad U time:")
cur_time = time.time()
myGradU(X, U, V).block_until_ready()
print(time.time() - cur_time)
# %timeit myGradU(X, U, V).block_until_ready()

print("autograd U time:")
cur_time = time.time()
jit_gradU(X, U, V).block_until_ready()
print(time.time() - cur_time)
# %timeit jit_gradU(X, U, V).block_until_ready() #strange...

print("My grad V time:")
cur_time = time.time()
myGradV(X, U, V).block_until_ready()
print(time.time() - cur_time)
# %timeit myGradV(X, U, V).block_until_ready()

print("autograd V time:")
cur_time = time.time()
jit_gradV(X, U, V).block_until_ready()
print(time.time() - cur_time)
# %timeit jit_gradV(X, U, V).block_until_ready() 
````

$$ R(x) = \frac{(Ax, x)}{(x, x)}, $$

$d(R(x)) = d(\frac{(Ax, x)}{(x, x)}) = \frac{d(Ax,x)(x,x) - (Ax,x)d(x,x)}{(x,x)^2} = \frac{2(Ax,dx)(x,x) - 2(Ax,x)(x,dx)}{(x,x)^2} =$

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(DEVICE), target.to(DEVICE)
            output = model(data.cuda())
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    return 100. * correct / len(test_loader.dataset)
````

### Create a new model, which is an analogue to the class ```Net```

````python
class myLinear(nn.Module):
    def __init__(self, weights, bias, rank):
      super(myLinear, self).__init__()
      
      U, S, V = torch.svd(weights)
      V = V.T

      U = U[:,:rank] 
      S = S[:rank]
      V = V[:rank, :]
      
      self.U = nn.Parameter(U)
      self.S = nn.Parameter(S)
      self.V = nn.Parameter(V)

      self.bias = nn.Parameter(torch.Tensor(bias.cpu()))

    def forward(self, input):
      weights = self.U @ torch.diag(self.S) @ self.V
      return input @ weights.T + self.bias

class myNet(nn.Module):
    def __init__(self, model, r):
      super(myNet, self).__init__()
      
      self.ReLU = nn.ReLU()

      params = model.state_dict()
      self.fc1 = myLinear(params['fc1.weight'], params['fc1.bias'], r)
      self.fc2 = myLinear(params['fc2.weight'], params['fc2.bias'], r)
      self.fc3 = myLinear(params['fc3.weight'], params['fc3.bias'], r)
      self.fc4 = myLinear(params['fc4.weight'], params['fc4.bias'], r)
      self.fc5 = myLinear(params['fc5.weight'], params['fc5.bias'], r)
      self.fc6 = myLinear(params['fc6.weight'], params['fc6.bias'], r)


    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2(x)
        x = self.ReLU(x)
        x = self.fc3(x)
        x = self.ReLU(x)
        x = self.fc4(x)
        x = self.ReLU(x)
        x = self.fc5(x)
        x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)
````
